## Global Hunger Index (GHI) Scores over Time

```python
data.replace("<5", 4, inplace=True)
```

```python
data['2000'] = data['2000'].astype(float)
data['2006'] = data['2006'].astype(float)
data['2012'] = data['2012'].astype(float)
data['2021'] = data['2021'].astype(float)
```

```python
cols_to_fill = ['2000', '2006', '2012', '2021']
data[cols_to_fill] = data[cols_to_fill].apply(lambda row: row.fillna(row.mean()), axis=1)
```

```python
filtered_df = data.copy()
numeric_columns = ['2000', '2006', '2012', '2021']
filtered_df[numeric_columns] = filtered_df[numeric_columns].apply(pd.to_numeric, errors='coerce')


filtered_df.head(10)
```

#### Analysis

###### Analysis of Each Country

```python
# Create a copy of the filtered DataFrame to work with GHI scores only
ghi_df = filtered_df[['Country', '2000', '2006', '2012', '2021']].copy()

# Melt the DataFrame to gather all years into a single column
ghi_df = ghi_df.melt(id_vars='Country', var_name='Year', value_name='GHI Score')

# Convert 'Year' column to numeric data type
ghi_df['Year'] = ghi_df['Year'].astype(int)

# Line plot for GHI Scores over time for each country
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='GHI Score', hue='Country', data=ghi_df)
plt.title('Global Hunger Index (GHI) Scores over Time for Each Country')
plt.xlabel('Year')
plt.ylabel('GHI Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

###### Analysis of Each Continent

```python
years = ['2000', '2006', '2012', '2021']
filtered_df[years] = filtered_df[years].apply(pd.to_numeric)

# Group the DataFrame by 'Continent' and calculate the mean values for each year
continent_mean_df = filtered_df.groupby('Continent')[years].mean()

# Transpose the DataFrame to get years as columns and continents as rows
continent_mean_df = continent_mean_df.T
plt.figure(figsize=(6, 4))
for continent in continent_mean_df.columns:
    plt.plot(years, continent_mean_df[continent], label=continent)
    

plt.title('Global Hunger Index (GHI) Scores')
plt.xlabel('Year')
plt.ylabel('GHI Score')
plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(False)
plt.show()
```
